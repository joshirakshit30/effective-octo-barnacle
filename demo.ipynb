```python
img = cv2.imread('triangle.jpg')

#bgr to rgb
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#bgr to gray
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

fig, plots = plt.subplots(1,2,figsize= (12,10))

plots[0].imshow(image)

plots[1].imshow(gray, cmap='gray')
```

```python
fast = cv2.FastFeatureDetector_create(10)

#features with nonMaxSupression
keypoint_with_supression = fast.detect(gray, None)

image_wnms = cv2.drawKeypoints(image, keypoint_with_supression,None, color= (0,0,255))

print(len(keypoint_with_supression))

fig, plots = plt.subplots(1,2,figsize=(14,10))

plots[0].imshow(image)

plots[1].imshow(image_wnms)
```

```python
image.shape
```

```python
pts = cv2.KeyPoint_convert(keypoint_with_supression)
dataframe = pd.DataFrame(pts)
dataframe[0].max()
```

```python
seq = np.arange(0,256,1)
seq2 = np.arange(0,255,2)
X,Y = np.meshgrid(seq,seq)

axis1,axis2 = np.meshgrid(seq2,seq2)


for j in range(256):
  X[j, 255] =0
  

fig, plots = plt.subplots(1,2, figsize = (14,11))

plots[0].imshow(X, cmap='gray')

plots[1].imshow(axis1, cmap='gray')
```

# Blurring Images

```python
filter = np.ones((5,5), np.float32)/32

horseImg = cv2.imread('running_horse.jfif')
horseImg = cv2.cvtColor(horseImg, cv2.COLOR_BGR2RGB)

horseImg.shape

blur = cv2.filter2D(horseImg, -1, filter)

fig, plots = plt.subplots(1,2, figsize = (14,12))

plots[0].imshow(horseImg)
plots[1].imshow(blur)
```

```python
horseImg.shape
```

```python
gaussianBlur = cv2.GaussianBlur(horseImg, (5,5),0)
fig, plots = plt.subplots(1,3, figsize=(14,10))

plots[0].imshow(horseImg)
plots[1].imshow(blur)
plots[2].imshow(gaussianBlur)
```
